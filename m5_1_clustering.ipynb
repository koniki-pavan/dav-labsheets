```python
import pandas as pd
df=pd.read_csv("/content/Mall_Customers (1).csv")
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
X=df[['Age','Spending Score (1-100)']].copy()
```

```python
for i in range(1,11):
  kmeans=KMeans(n_clusters=i, init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(X)
```

wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i, init='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('Selecting the number of Clusters using the Elbow Method')
plt.xlabel('Clusters')
plt.ylabel("WCSS")
plt.show()
```

```python
for k in range(1,11):
  plt.scatter(X["Age"],X["Spending Score (1-100)"])
  plt.title("Clusters Identified by K-Means Clustering")
plt.ylabel("Spending Score (1-100)")
plt.xlabel("Age")
plt.show()
```
