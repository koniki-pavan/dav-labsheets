```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/netflix_titles.csv',usecols=['cast'])
```

```python
df.head()
```

```python
ndf=df.dropna()
ndf.head()
```

```python
text = " ".join(item for item in ndf['cast'])
print(text)
```

```python
  stopwords = set(STOPWORDS)
```

```python
wordcloud=WordCloud(background_color='linen').generate(text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.margins(x=0,y=0)
plt.show()
```

```python
wordcloud=WordCloud(background_color='linen',
                    max_words=100,
                    max_font_size=300,
                    width=800,
                    height=500,
                    colormap='cool').generate(text)
plt.figure(figsize=(20,20))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.margins(x=0,y=0)
plt.savefig('cloud.jpg',format='jpg')
plt.show()
```
